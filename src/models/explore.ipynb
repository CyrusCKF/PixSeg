```python
import numpy as np
import torch
from torch import Tensor
import torchinfo
from torchvision.models import segmentation
```

```python
model = segmentation.deeplabv3_mobilenet_v3_large(num_classes=10)
fake_input = torch.rand([4, 3, 224, 224], dtype=torch.float)
fake_output: dict[str, Tensor] = model(fake_input)
for k, v in fake_output.items():
    print(k, v.dtype, v.shape, v.min().item(), v.max().item())
```

```python
torchinfo.summary(
    model,
    input_data=fake_input,
    col_names=("input_size", "output_size", "num_params"),
)
```
